```python
folder_path = '/content/drive/MyDrive/All Chemicals/'
line_count_dict = {}
total = 0

for root, dirs, files in os.walk(folder_path):
    for file_name in files:
        if file_name.endswith('.txt') and root != folder_path:
            file_path = os.path.join(root, file_name)
            with open(file_path,'r') as file:
                line_count = sum(1 for line in file)
            line_count = max(line_count - 1, 0)
            print(file_name, line_count)
            line_count_dict[file_path] = line_count
            total = total + line_count
```

```python
folder_path = '/content/drive/MyDrive/All Chemicals/'
SELFIES_list = []
SMILES_list = []

for root, dirs, files in os.walk(folder_path):
    for file_name in files:
        if file_name.endswith('.txt') and root != folder_path:
            file_path = os.path.join(root, file_name)
            with open(file_path, 'r') as input_file:
                lines = input_file.readlines()
                if lines:
                    lines = lines[1:] if lines[0].strip() else lines
                    for line in lines:
                        line_content = line.split('\t', 1)[0]
                        try:
                            SELFIES_list.append(sf.encoder(line_content))
                            SMILES_list.append(line_content)
                        except:
                            pass
```
